```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib widget
```

```python
# Load in csv
pyber_ride_df = pd.read_csv("../Resources/avg_rain_state.csv")
pyber_ride_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.subplots(figsize=(10, 5))
plt.bar(x_axis, pyber_ride_df["Inches"])
plt.xticks(tick_locations, pyber_ride_df["State"])
plt.show()
```

```python
# Set x and y limits
plt.subplots(figsize=(10, 5))
```

```python
# Set a Title and labels
plt.bar(x_axis, pyber_ride_df["Inches"], color="skyblue", label="Inches")

```
