```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib widget
```

```python
# Read CSV
settlements = pd.read_csv("Resources/union_settlements_1995.csv")
settlements.head()
```

```python
# List columns
settlements.columns
```

```python
# Get total settlements by union
settlements_by_union = settlements.groupby('COMPANY IDENTIFYING INFORMATION ').sum()
settlements_by_union.head()
```

```python
# Configure plot, figsize, title, and axis labels
settlements_by_union.plot(kind="bar", figsize=(20,3))
plt.title("Total Settlements by Union")
plt.ylabel("Total Settlements")
plt.xlabel("Union")


# Configure x-tick rotation
plt.xticks(rotation=45)

# Show plot
plt.show()
```

```python
# Resize plot to display labels
plt.figure(figsize=(10, 6))
plt.subplots_adjust(bottom=0.2)

# Plot the data
settlements_by_union.plot(kind="bar")

# Configure plot title and axis labels
plt.title("Total Settlements by Union")
plt.xlabel("Union")
plt.ylabel("Total Settlements")

# Rotate x-tick labels
plt.xticks(rotation=45)

# Show the plot
plt.show()

```
