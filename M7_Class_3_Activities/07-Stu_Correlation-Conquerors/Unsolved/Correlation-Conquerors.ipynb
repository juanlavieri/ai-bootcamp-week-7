```python
# Dependencies
import pandas as pd
import sklearn.datasets as dta
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
# Read in the wine recognition data set from sklearn and load into Pandas
data = dta.load_wine()
wine_data = pd.DataFrame(data.data,columns=data.feature_names)
print(data.DESCR)
```

```python
# Plot flavanoids versus malic_acid on a scatterplot
plt.scatter(wine_data.flavanoids,wine_data.malic_acid)
plt.xlabel('Flavanoids')
plt.ylabel('Malic Acid')
plt.show()

```

```python
# Calculate the correlation coefficient between malic_acid and flavanoids
correlation = st.pearsonr(wine_data.flavanoids,wine_data.malic_acid)
print(f"The correlation between both factors is {round(correlation[0],2)}")

```

```python
# Plot colour_intensity versus alcohol on a scatterplot
plt.scatter(wine_data.color_intensity,wine_data.alcohol)
plt.xlabel('Color Intensity')
plt.ylabel('Alcohol')
plt.show()
```

```python
# Calculate the correlation coefficient between alcohol and color_intensity
correlation = st.pearsonr(wine_data.color_intensity,wine_data.alcohol)
print(f"The correlation between both factors is {round(correlation[0],2)}")

```

```python
corr_matrix = wine_data.corr()
strongest_positive = corr_matrix.unstack().sort_values(ascending=False).drop_duplicates().head(10)
strongest_negative = corr_matrix.unstack().sort_values().drop_duplicates().head(10)
strongest_correlations = pd.concat([strongest_positive, strongest_negative])
print(strongest_correlations)
```
